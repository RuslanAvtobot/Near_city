```python
import pandas as pd
import numpy as np
import math
```

```python
from chardet.universaldetector import UniversalDetector
detector = UniversalDetector()
with open('cities.csv', 'rb') as fh:
    for line in fh:
        detector.feed(line)
        if detector.done:
            break
    detector.close()
print(detector.result)
```

```python
cities['Часовой пояс'] = cities['Часовой пояс'].str.replace('UTC', '') #удаляю UTC из часового пояса
```

```python
cities['Часовой пояс'] = cities['Часовой пояс'].apply(lambda x: int(x))
```

```python
int(cities[cities['Город'] == 'Казань']['Часовой пояс'].values)+2.5
```

```python
#Отсекаем часовые пояса +- 2 часа
near_Kazan = cities[cities['Часовой пояс'].between(float(cities[cities['Город'] == 'Казань']['Часовой пояс'].values)-2.5,
                                            float(cities[cities['Город'] == 'Казань']['Часовой пояс'].values)+2.5,
                                            inclusive = True)]
```

```python
# Отсекаем города по Долготе
near_Kazan = near_Kazan[near_Kazan['Долгота'].between(float(near_Kazan[near_Kazan['Город'] == 'Казань']['Долгота'].values)-1,
                                            float(near_Kazan[near_Kazan['Город'] == 'Казань']['Долгота'].values)+1,
                                            inclusive = True)]
```

```python
max_latitude = 40000/360 * math.cos(float(near_Kazan[near_Kazan['Город'] == 'Казань']['Широта'].values))
```

```python
latitude = round(111.111/(max_latitude/60)/60, 6)
```

```python
near_Kazan = near_Kazan[near_Kazan['Широта'].between(float(near_Kazan[near_Kazan['Город'] == 'Казань']
                                                           ['Широта'].values)-latitude,
                                                     float(near_Kazan[near_Kazan['Город'] == 'Казань']
                                                           ['Широта'].values)+latitude,
                                            inclusive = True)]
```
